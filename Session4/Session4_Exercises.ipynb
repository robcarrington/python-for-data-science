## I. For Loops


### Warm Ups
---
**For-loop on a list:** 
    
```python
for num in [1, 2, 3]:
    print(num)
```

**For-loop on a string:**

```python
for char in "word":
    print(char)
```

**For-loop on a range:**
    
```python
for num in range(10):
    print(num)

```

**Using a for-loop to construct:**
    
```python
i = 0
for num in range(10):
    i = i + num
```

## II. Dictionaries


### Warm Ups
---

**Dictionary literal:** `my_dict = {"name": "Rob", "shoe_size": 9.5}`

**Fetch by key:** `my_dict['name']`

**Add an item:** `my_dict['hair'] = "brown"`

**Looping over a dictionary:** 

```python
for key in my_dict:
    print(key)
```

**Looping over key-value pairs:** 

```python
for key,value in my_dict.items():
    print(key)
    print(value)
```
